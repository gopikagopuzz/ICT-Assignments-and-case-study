## ðŸ§  **Question 1: NumPy Array Manipulation**
1. Create a 1D NumPy array with integers from **1 to 24**.  
2. Reshape the array into a **4x6 matrix**.  
3. Extract the **third row** and **second column** from the reshaped array.  
4. Transpose the matrix and display the result.

```python
# Solution for Question 1
import numpy as np

# Your code here

print('1D NumPy array with integers from 1 to 24.')
list_1 = list(range(1,25))
print(list_1)
array_1 = np.array(list_1)
print(array_1)
#Reshape the array into a 4x6 matrix.
print('Reshape the array into a 4x6 matrix.')
array_2 = array_1.reshape(4,6)
print(array_2)

#Extract the third row and second column from the reshaped array.
print('third row is:')
thrdr =array_2[2,:]
print(thrdr)
print('second column is:')
scdc =array_2[:,1]
print(scdc)

#Transpose the matrix and display the result.
print('transposed matrix is:')
array_3=array_2.T
array_3
```

## ðŸ§  **Question 2: Array Element Extraction**
1. Create a **5x5 NumPy array** with random integers between **10 and 100**.  
2. Extract all the **even numbers** from the array.  
3. Replace all numbers **greater than 50** with **0**.  
4. Extract the diagonal elements of the array.

```python
# Solution for Question 2
import numpy as np

# Your code here
#Create a 5x5 NumPy array with random integers between 10 and 100.
print('5x5 NumPy array with random integers between 10 and 100.')
rndarr =np.random.randint(10,101,(5,5))
print(rndarr)

#Extract all the even numbers from the array.
print('all the even numbers from the array are:')
even=rndarr[rndarr % 2==0]
print(even)


#Replace all numbers greater than 50 with 0.
print('repaced array is:')
rndarr[rndarr>50]=0
print(rndarr)

#Extract the diagonal elements of the array.
print('The diagonal elements of the array are: ')
diag =np.diag(rndarr)
print(diag)
```

## ðŸ§  **Question 3: DataFrame Creation and Operations (Pandas)**
1. Create a Pandas DataFrame containing the following data:

| Student_ID | Name   | Subject | Marks |
|------------|--------|---------|-------|
| 1          | Alice  | Math    | 88    |
| 2          | Bob    | Science | 75    |
| 3          | Carol  | History | 92    |
| 4          | David  | Math    | 85    |
| 5          | Eva    | Science | 90    |

2. Find the **average marks** for each subject.  
3. Add a new column named **Grade** based on the marks:
- Marks â‰¥ 90 â†’ A
- Marks â‰¥ 80 â†’ B
- Marks â‰¥ 70 â†’ C
- Else â†’ D

```python
# Solution for Question 3
import pandas as pd

# Your code here
#Create a Pandas DataFrame containing the following data:
data = {'Student_ID':[1,2,3,4,5],'Name':['Alice','Bob','Carol','David','Eva'],'Subject':['Math','Science','History','Math','Science'],'Marks':[88,75,92,85,90]}
df =pd.DataFrame(data)
print(df)

#Find the average marks for each subject.
print('average marks for each subject.')
avg_marks =df.groupby('Subject')['Marks'].mean()
print(avg_marks)

#Add a new column named Grade based on the marks:
print('new column named Grade.')
def assign_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    else:
        return "D"

# Add the new column
df["Grade"] = df["Marks"].apply(assign_grade)

print(df)





```

## ðŸ§  **Question 4: Dataset Analysis (Pandas, Matplotlib, Seaborn)**
Use the dataset `diabetes.csv` and perform the following tasks:

1. Display the **BMI** and **Blood Pressure** for samples with **index ranging from 20 to 40**.  
2. Find how many patients have **BMI > 30**.  
3. Find the **mean and standard deviation** of the **Glucose** levels.  
4. Plot a **boxplot** for the column **Insulin**.  
5. Create a **heatmap** showing the correlation between the numerical columns in the dataset.

```python
# Solution for Question 4
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Your code here
diabetes = pd.read_csv('/content/diabetes_model (1).csv')
#print(diabetes.head())

#Display the BMI and Blood Pressure for samples with index ranging from 20 to 40.
print('BMI and Blood Pressure for samples with index ranging from 20 to 40.')
bmi_bp =diabetes.loc[20:40,['BMI','BloodPressure']]
print(bmi_bp)


#Find how many patients have BMI > 30.
high = diabetes[diabetes['BMI'] > 30]['BMI'].count()
print('patients have BMI > 30 =',high)


#Find the mean and standard deviation of the Glucose levels.
mean = diabetes['Glucose'].mean()
std = diabetes['Glucose'].std()
print('mean of Glucose level =',mean)
print('std of Glucose level =',std)




```

```python
print('The box plot for insuline is:')
plt.boxplot(diabetes['Insulin'])
plt.title('Boxplot of Insulin Levels') # Add a title for better visualization
plt.ylabel('Insulin') # Add a y-axis label
plt.show()



```

```python
# heatmap showing the correlation between the numerical columns in the dataset.
print(' heatmap showing the correlation between the numerical columns in the dataset is:')
corr_matrix = diabetes.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
